# Advent of Code Day 6

## Part 1

```python
with open("Day6_input_1.txt", 'r') as f:
    initial_state = f.readline().split(sep=',')
```

```python
for i in range(len(initial_state)):
    initial_state[i] = int(initial_state[i])
```

```python
num_days = 80
cur_state = initial_state.copy()
for day in range(num_days):
    temp_state = []
    tot_zeros = 0
    for ele in cur_state:
        if ele == 0 :
            tot_zeros += 1
            temp_state.append(6)
        else:
            temp_state.append(ele-1)
    for i in range(tot_zeros):
        temp_state.append(8)
    cur_state = temp_state.copy()
```

```python
print("The answer is",len(cur_state))
```

## Part 2

```python
tracker = {}
for i in range(9):
    tracker[i] = 0
for ele in initial_state:
        tracker[ele] += 1
```

```python
def sum_fish(tracker):
    sum_fish_val = 0
    for key in tracker:
        sum_fish_val += tracker[key]
    return sum_fish_val
```

```python
def update_one_day(tracker):
    new_tracker = tracker.copy()
#     print(new_tracker)
    for key in new_tracker:
        if key in range(0,6):
            new_tracker[key] = tracker[key+1]
        else:
            new_tracker[7] = tracker[8]
            new_tracker[6] = tracker[0] + tracker[7]
            new_tracker[8] = tracker[0]
    return new_tracker
```

```python
updated_tracker = tracker.copy()
for i in range(256):
    print("Day:",i)
    updated_tracker = update_one_day(updated_tracker).copy()
    print(updated_tracker)
```

```python
print("The answer is:",sum_fish(updated_tracker))
```
